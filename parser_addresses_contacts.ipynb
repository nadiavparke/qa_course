```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from time import sleep
import pandas as pd
```

```python
url_address = "https://www.prepostseo.com/tool/ru/fake-address-generator"
```

```python
address_list = []
contact_list = []
countries = ["Finland", "France", "Germany", "Lithuania", "USA"]
```

```python
address_headers = ["country", "city", "street", "house"]
contact_headers = ['email', 'phone']
header_id = pd.Series(range(1,36))
person_id_header = pd.concat([pd.Series(range(1,31)),pd.Series(range(1,5))]).sample(frac=1).reset_index(drop=True)
```

```python
def create_profile(country):
        driver.execute_script(f"window.scrollTo(0, 300)") 
        sleep(2)
        country_button = driver.find_element(By.CSS_SELECTOR, f'[data-title="{country}"]')
        country_button.click()
        gen_button = driver.find_element(By.ID, 'calcRatio')
        gen_button.click()
        sleep(3)
```

```python
def add_address():
    country = driver.find_element(By.ID, 'proCountry').text
    city = driver.find_element(By.ID, 'proCity').text
    address = driver.find_element(By.ID, 'proStreetAddress').text.split()
    
    for i in range(len(address)):
        address[i] = address[i].translate({ord(i): None for i in ',./'})
        
    print(address)
        
    if address[-1].replace('-', '').isalpha() == False:
        if address[0].replace('-', '').isalpha() == False: #building - street - appartment
            address_list.append([country,city,' '.join(address[1:-1]).replace('Apt.',''), address[0].translate({ord(i): None for i in 'abcdef'})])
        elif address[-2].translate({ord(i): None for i in '-,./'}).isalpha() == False: #street - building - appartment
            address_list.append([country,city,' '.join(address[:-2]), address[-2].translate({ord(i): None for i in 'abcdef'})])
        else:#street - building
            address_list.append([country,city,' '.join(address[:-1]), address[-1].translate({ord(i): None for i in 'abcdef'})])    
    elif address[0].translate({ord(i): None for i in '-,./'}).isalpha() == False: #building - street
        address_list.append([country,city,' '.join(address[1:]), address[0].translate({ord(i): None for i in 'abcdef'})])
    else:#street
        address_list.append([country,city,' '.join(address), ])
            
```

```python
def add_contacts():
    email = driver.find_element(By.ID, 'proEmail').text
    phone = driver.find_element(By.ID, 'proPhone').text
    contact_list.append([email, phone])
```

```python
driver_service = Service(executable_path = "/home/nadego/Documents/chromedriver/")
driver = webdriver.Chrome(service = driver_service)
driver.get(url_address)
```

```python
for country in countries:    
    for i in range(7):
        create_profile(country)
        add_address()
        add_contacts()
```

```python
address_table = pd.DataFrame(address_list, columns=address_headers)
address_table['address_id'] = header_id
address_table = pd.DataFrame(address_table, columns=['address_id', "country", "city", "street", "house"])
address_table
```

```python
address_table.to_csv('address_table', index=False)
```

```python
contact_table = pd.DataFrame(contact_list, columns = contact_headers)
contact_table['contact_id'] = header_id
contact_table['persons_id'] = person_id_header
contact_table = pd.DataFrame(contact_table, columns = ['contact_id', 'email', 'phone', 'persons_id'])
contact_table
```

```python
contact_table.to_csv('contact_table', index=False)
```
